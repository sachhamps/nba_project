```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.kernel_ridge import KernelRidge
```

```python

x = np.linspace(0,5,100)
E = np.random.normal(0,1.3,[200,1])
W = np.array([1.3,0.5])
Y = np.array(W[0]*x + W[1])

for i in range (0,len(Y)):
    Y[i] = Y[i] + E[i]

plt.figure(1)
sns.scatterplot(x,Y,alpha=1)
plt.savefig('scatterplt_ex.png')
plt.figure(2)

sns.scatterplot(x,Y,alpha=1)
clf = LinearRegression()
x = x.reshape(-1,1)
Y = Y.reshape(-1,1)
clf_fit = clf.fit(x,Y)

p = clf_fit.predict(x)

plt.plot(x,p,'magenta')
plt.savefig('linregscatterplt.png')
plt.show()
```

```python
X = np.linspace(-np.pi,np.pi,50)
# X = X.reshape(-1,1)

E = np.random.normal(0,0.3,[100,1])
Y = np.sin(X)

for i in range (len(Y)):
    Y[i] = Y[i] + E[i]
    
sns.scatterplot(X,Y,alpha=0.9)
plt.savefig('wavey.png')
plt.show()
```

```python
lr = LinearRegression()
X = X.reshape(-1,1)
fit = lr.fit(X,Y)
p = fit.predict(X)
X = X.flatten()
sns.scatterplot(X,Y)
plt.plot(X,p,'magenta')
plt.savefig('wavey_lr.png')
plt.show()
```

```python
k = KernelRidge(alpha=0.5, kernel='rbf', gamma=0.7, degree=3, coef0=1, kernel_params=None)
X = X.reshape(-1,1)
f = k.fit(X,Y)
p = f.predict(X)
X = X.flatten()
sns.scatterplot(X,Y)
plt.plot(X,p,'lime')
plt.savefig('wavey_kr.png')
plt.show()
```
